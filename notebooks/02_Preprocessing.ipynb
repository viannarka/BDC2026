## Impoprt Libraries

```python
import torch
from torchvision import transforms
```

### Resize the image to 224x224 pixels and Normalize the pixel values to be between 0 and 1.

### 2. Membuat Pipeline Preprocessing

Mendefinisikan urutan transformasi yang akan diterapkan pada setiap gambar sebelum digunakan oleh model.

```python
train_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])
```

### 3. Membaca Contoh Gambar

Membuka salah satu gambar dari dataset sebagai sampel untuk menguji pipeline preprocessing.

```python
from PIL import Image

image = Image.open("../data/train/0_Recyclable/R_23.jpg")
image
```

### 4. Menerapkan Pipeline Preprocessing

Mengubah gambar menjadi tensor menggunakan pipeline yang telah dibuat.

```python
tensor = train_transform(image)
```

### 5. Memeriksa Hasil Transformasi

Memastikan bahwa gambar telah berhasil diubah menjadi tensor dengan ukuran dan rentang nilai yang sesuai.

```python
print(type(tensor))
print(tensor.shape)
print(tensor.min())
print(tensor.max())
```

### 6. Melihat Nilai RGB pada Satu Pixel

Mengambil nilai RGB dari satu pixel untuk memahami bagaimana gambar direpresentasikan sebagai tensor.

```python
print(tensor[:, 0, 0])
```
